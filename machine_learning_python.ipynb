# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Poor, Standard, Good


```python
### Score de crédito do cliente
## Boa - Good
## OK - Standard
## Ruim - Poor


### Passo 0: Entender o desafio e a empresa
## Precisamos prever a nota de crédito do cliente


### Passo 1: Importar a base de dados
import pandas as pd # Trabalha com banco de dados

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
### Passo 2: Preparar a base de dados para a Inteligência Artificial
## Todas as informações precisam ser transformadas em numéricas para que a IA consiga entender (exceto a informação que eu quero prever)
display(tabela.info()) # Identificar os tipos de informações da tabela que estão como caracteres


##LabelEncoder
import sklearn # Pacote que permite criar as IAs
from sklearn.preprocessing import LabelEncoder # Importar apenas a função que vamos usar para transformar as informações em numéricas

# profissao
codificador1 = LabelEncoder() # É preciso criar um codificador para cada informação que vamos alterar, pois os mesmos números em cada coluna significam coisas distintas
tabela["profissao"] = codificador1.fit_transform(tabela["profissao"])

# mix_credito
codificador2 = LabelEncoder()
tabela["mix_credito"] = codificador2.fit_transform(tabela["mix_credito"])

# comportamento_pagamento
codificador3 = LabelEncoder()
tabela["comportamento_pagamento"] = codificador3.fit_transform(tabela["comportamento_pagamento"])

display(tabela.info())
```

```python
### Passo 2.5: Separar as informações da base de dados para a IA
## Separar em X(informações que utilizamos para tentar prever) e Y(informação que queremos prever)

y = tabela["score_credito"]
x = tabela.drop(columns=["score_credito", "id_cliente"])

## Separar em dados de treino e de teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3) # A ordem necessariamente tem que ser essa na hora de fazer a separação
# O "test_size = 0.3" significa que foi separado 30% do banco de dados para a IA utilizar como teste
```

```python
### Passo 3: Criar o modelo de IA capaz de prever a nota de crédito do cliente
## Tipos de modelo:
# Árvore de Decisão -> RandomForest
# Vizinhos Próximos -> KNN

## Importar o modelo
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

## Criar o modelo
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

## Treinar o modelo
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
### Passo 4: Escolher o melhor modelo
## Previsões
previsao_arvorededecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

## Acuracia
from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvorededecisao))
display(accuracy_score(y_teste, previsao_knn))
```

```python
### Passo 4 resposta:
## O melhor modelo é de Árvore de Decisão, pois teve a melhor acuracia entre o gabarito e o teste
```

```python
### Passo 5: Fazer novas previsões
## Importar novos dados
tabela_nova = pd.read_csv("novos_clientes.csv")
display(tabela_nova)

## É preciso repetir a transformação dos tipos de dados nos novos dados inseridos, utilizando os codificadores já criados anteriormente

# profissao
tabela_nova["profissao"] = codificador1.fit_transform(tabela_nova["profissao"])

# mix_credito
tabela_nova["mix_credito"] = codificador2.fit_transform(tabela_nova["mix_credito"])

# comportamento_pagamento
tabela_nova["comportamento_pagamento"] = codificador3.fit_transform(tabela_nova["comportamento_pagamento"])

## Nova previsão com os novos dados
previsao = modelo_arvoredecisao.predict(tabela_nova)
display(previsao)
```
